```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import FancyArrowPatch
import warnings
from gensim.models import Word2Vec
import re
from gensim.models import FastText
from sentence_transformers import SentenceTransformer
from sklearn.feature_extraction.text import TfidfVectorizer
# Calcular similitud coseno entre pares
from sklearn.metrics.pairwise import cosine_similarity
```

```python
corpus_profesor = [
    ["el", "procesamiento", "del", "lenguaje", "natural", "es", "una", "rama", "de", "la", "inteligencia", "artificial"],
    ["los", "modelos", "de", "embeddings", "como", "word2vec", "capturan", "similitud", "semántica"],
    ["en", "esta", "clase", "de", "nlp", "aprendemos", "a", "procesar", "texto"],
    ["la", "inteligencia", "artificial", "utiliza", "aprendizaje", "profundo", "y", "modelos", "de", "lenguaje"],
    ["el", "profesor", "explica", "conceptos", "de", "procesamiento", "de", "lenguaje", "natural"]
]

modelo_w2v = Word2Vec(
    sentences=corpus_profesor, 
    vector_size=100,  
    window=5,         
    min_count=1,      
    workers=4
)

print("¡Modelo Word2Vec entrenado con éxito!")
```

```python
palabra_clave = 'inteligencia'

top_5_similares = modelo_w2v.wv.most_similar(palabra_clave, topn=5)

print(f"--- 5 Palabras más similares a '{palabra_clave}' ---")
for palabra, score in top_5_similares:
    print(f"• {palabra}: {score:.4f} (Similitud)")

print("\n" + "-"*40 + "\n")

try:
    analogia = modelo_w2v.wv.most_similar(positive=['modelos', 'explicar'], negative=['profesor'], topn=1)
    print(f"--- Resultado de la analogía ---")
    print(f"profesor -> explicar | modelos -> {analogia[0][0]}")
except KeyError as e:
    print("Nota: Algunas palabras de la analogía no alcanzaron a procesarse por el tamaño del corpus.")
```

```python
from sklearn.decomposition import PCA

palabra_clave = 'inteligencia'
top_5_similares = modelo_w2v.wv.most_similar(palabra_clave, topn=5)

palabras_a_graficar = [palabra_clave] + [palabra for palabra, _ in top_5_similares]
vectores = [modelo_w2v.wv[word] for word in palabras_a_graficar]

pca = PCA(n_components=2)
vectores_2d = pca.fit_transform(vectores)

plt.figure(figsize=(9, 7))
plt.grid(True, linestyle='--', alpha=0.5)

colores = ['red'] + ['blue'] * 5

plt.scatter(vectores_2d[:, 0], vectores_2d[:, 1], c=colores, s=120, edgecolors='k')

for i, palabra in enumerate(palabras_a_graficar):
    if palabra == palabra_clave:
        plt.text(vectores_2d[i, 0] + 0.005, vectores_2d[i, 1] + 0.005, palabra.upper(), 
                 fontsize=12, weight='bold', color='red')
    else:
        plt.text(vectores_2d[i, 0] + 0.005, vectores_2d[i, 1] + 0.005, palabra, 
                 fontsize=11, weight='medium')

plt.title(f"PCA: '{palabra_clave}' y sus 5 palabras más similares", fontsize=14, pad=15)
plt.xlabel("Componente Principal 1")
plt.ylabel("Componente Principal 2")

plt.show()
```

```python
def buscador_semantico(query, modelo, top_n=3):

    if query not in modelo.wv:
        return f"La palabra '{query}' no se encuentra en el vocabulario del modelo."

    vector_query = modelo.wv[query].reshape(1, -1)
    
    resultados = []
    

    for palabra in modelo.wv.index_to_key:
        if palabra == query:
            continue  
            
        vector_palabra = modelo.wv[palabra].reshape(1, -1)
        
 
        similitud = cosine_similarity(vector_query, vector_palabra)[0][0]
        resultados.append((palabra, similitud))
    

    resultados.sort(key=lambda x: x[1], reverse=True)
    
    return resultados[:top_n]

palabra_buscada = 'procesamiento'
mejores_resultados = buscador_semantico(palabra_buscada, modelo_w2v, top_n=3)

print(f"🔍 Resultados de búsqueda semántica para: '{palabra_buscada}'")
print("-" * 50)
for rango, (palabra, score) in enumerate(mejores_resultados, 1):
    print(f"{rango}. Palabra: '{palabra}' | Similitud Coseno: {score:.4f}")
```
