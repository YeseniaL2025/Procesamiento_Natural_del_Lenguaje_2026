```python
import numpy as np
import pandas as pd
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.tokenize import TweetTokenizer
from sklearn.model_selection import train_test_split

# Descargar recursos necesarios de NLTK
nltk.download('stopwords')

# 1. Cargar el dataset
df = pd.read_csv('Data/moviereviews.tsv', delimiter='\t')

# 2. Limpieza de nulos y mapeo de etiquetas a formato numérico (0, 1)
df.dropna(subset=['review', 'label'], inplace=True)
df['target'] = df['label'].map({'neg': 0, 'pos': 1})

print(f"Total de reseñas cargadas: {len(df)}")
df.head()
```

```python
def process_tweet(tweet):
    """Tokeniza, remueve stopwords y aplica stemming a la reseña."""
    stemmer = PorterStemmer()
    stopwords_english = stopwords.words('english')

    # Limpiar texto
    tweet = re.sub(r'\$\w*', '', tweet)
    tweet = re.sub(r'^RT[\s]+', '', tweet)
    tweet = re.sub(r'https?://[^\s\n\r]+', '', tweet)
    tweet = re.sub(r'#', '', tweet)

    tokenizer = TweetTokenizer(
        preserve_case=False,
        strip_handles=True,
        reduce_len=True
    )

    tweet_tokens = tokenizer.tokenize(tweet)

    tweets_clean = []

    for word in tweet_tokens:
        if word not in stopwords_english and word.isalnum():
            stem_word = stemmer.stem(word)
            tweets_clean.append(stem_word)

    return tweets_clean


def build_freqs(tweets, ys):
    """
    Crea un diccionario:
    (palabra, etiqueta) -> frecuencia
    """
    yslist = np.squeeze(ys).tolist()

    freqs = {}

    for y, tweet in zip(yslist, tweets):
        for word in process_tweet(tweet):

            pair = (word, float(y))

            if pair in freqs:
                freqs[pair] += 1
            else:
                freqs[pair] = 1

    return freqs
```

```python
def extract_features(tweet, freqs):
    """
    Regresa un vector [1, frecuencia positiva, frecuencia negativa]
    """
    word_l = process_tweet(tweet)

    x = np.zeros((1, 3))

    # bias
    x[0, 0] = 1

    for word in word_l:
        x[0, 1] += freqs.get((word, 1.0), 0)
        x[0, 2] += freqs.get((word, 0.0), 0)

    return x
```

```python
# Textos y etiquetas
X_reviews = df['review'].tolist()
Y_labels = df['target'].values.reshape(-1,1)

# División entrenamiento-prueba
X_train_txt, X_test_txt, y_train, y_test = train_test_split(
    X_reviews,
    Y_labels,
    test_size=0.2,
    random_state=42,
    stratify=Y_labels
)

# Construir frecuencias
freqs = build_freqs(X_train_txt, y_train)

# Matriz entrenamiento
X_train = np.zeros((len(X_train_txt),3))

for i in range(len(X_train_txt)):
    X_train[i,:] = extract_features(X_train_txt[i], freqs)

# Matriz prueba
X_test = np.zeros((len(X_test_txt),3))

for i in range(len(X_test_txt)):
    X_test[i,:] = extract_features(X_test_txt[i], freqs)

print("Dimensiones de X_train:", X_train.shape)
print("Dimensiones de X_test:", X_test.shape)
```

```python
def sigmoid(z):
    """Calcula la función sigmoide."""
    z = np.clip(z, -500, 500) 
    return 1 / (1 + np.exp(-z))

def gradientDescent(x, y, theta, alpha, num_iters):
    """Ajuste iterativo de pesos mediante Gradiente Descendiente Batch."""
    m = x.shape[0]
    historial_costo = [] # LISTA PARA GUARDAR LOS COSTOS

    for i in range(0, num_iters):
        # Multiplicación matricial: Logits
        z = np.dot(x, theta)
        h = sigmoid(z)

        # Cálculo de la función de costo (agregando un épsilon 1e-15 para evitar log(0))
        J = -1./m * (np.dot(y.transpose(), np.log(h + 1e-15)) + np.dot((1-y).transpose(), np.log(1-h + 1e-15)))
        
        # Guardamos el costo de esta iteración extraído como escalar
        historial_costo.append(float(J.item()))

        # Actualizando los pesos del modelo (Gradiente)
        theta = theta - (alpha/m) * np.dot(x.transpose(), (h-y))

    return historial_costo, theta
```

```python
theta_inicial = np.zeros((3,1))

alpha = 1e-6
iterations = 1500

historial_loss, theta_optimizado = gradientDescent(
    X_train,
    y_train,
    theta_inicial,
    alpha,
    iterations
)

print("¡Modelo entrenado exitosamente!")
print("Costo final:", historial_loss[-1])
print("Theta óptimo:")
print(theta_optimizado)
```

```python
def predict_tweet(tweet, freqs, theta):

    x = extract_features(tweet, freqs)

    y_pred = sigmoid(np.dot(x, theta))

    return y_pred


def test_logistic_regression(test_x, test_y, freqs, theta):

    y_hat = []

    for tweet in test_x:

        y_pred = predict_tweet(tweet, freqs, theta)

        if y_pred > 0.5:
            y_hat.append(1)
        else:
            y_hat.append(0)

    y_hat = np.array(y_hat).reshape(-1,1)

    accuracy = (y_hat == test_y).sum()/len(test_x)

    return accuracy


accuracy_final = test_logistic_regression(
    X_test_txt,
    y_test,
    freqs,
    theta_optimizado
)

print(f"Accuracy final en el conjunto de prueba: {accuracy_final:.4f}")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.plot(historial_loss)
plt.xlabel('Iteraciones')
plt.ylabel('Costo (Loss)')
plt.title('Convergencia del Gradiente Descendente')
plt.grid(True)
plt.show()
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

# Predicciones
y_pred = []

for tweet in X_test_txt:
    pred = predict_tweet(tweet, freqs, theta_optimizado)

    if pred > 0.5:
        y_pred.append(1)
    else:
        y_pred.append(0)

# Matriz de confusión
cm = confusion_matrix(y_test, y_pred)

disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot(cmap='Blues')
plt.title('Matriz de Confusión')
plt.show()
```

```python
accuracy = accuracy_final * 100

plt.figure(figsize=(5,5))
plt.bar(['Accuracy'], [accuracy])
plt.ylim(0,100)
plt.ylabel('Porcentaje (%)')
plt.title('Exactitud del Modelo')
plt.show()
```
